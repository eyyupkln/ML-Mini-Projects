```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest (
     n_estimators=200,
    contamination=0.001727, #it is percentage of contamination ratio
    max_samples="auto",
    max_features=1.0,
    random_state=42,
    n_jobs=-1
)
```

```python
iso.fit(X)
```

```python
df["iso_raw"] = iso.predict(X)
```

```python
df['iso_raw'].unique()
```

```python
#1  → normal
#-1 → anomaly
```

```python
df["iso_pred"] = df["iso_raw"].map({1:0, -1:1})
```

```python
df["iso_pred"].unique()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(["Class","iso_raw","iso_pred"], axis=1)
y = df["Class"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    stratify=y,   # it will be important for isolation forest.
    random_state=42
)
```

```python
iso_test_pred = df.loc[X_test.index, "iso_pred"]
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, iso_test_pred))
```

```python
train_anomaly_scores = iso.decision_function(X_train)
test_anomaly_scores = iso.decision_function(X_test)
```

```python
X_train_hybrid = X_train.copy()
X_test_hybrid = X_test.copy()
```

```python
X_train_hybrid['Iso_Score'] = train_anomaly_scores
X_train_hybrid['Iso_Score'] = test_anomaly_scores
```

```python
df2=X_train_hybrid +X_train_hybrid
```

```python
df2['Class']=df['Class']
```

```python
sns.kdeplot(
    df2[df["Class"]==0]["Iso_Score"],
    label="Normal"
)

sns.kdeplot(
    df2[df["Class"]==1]["Iso_Score"],
    label="Fraud"
)

plt.legend()
plt.title("Anomaly Score Distribution")
plt.show()
```

```python
# let's see top 1% suspicious transaction
```

```python
threshold= df2["Iso_Score"].quantile(0.01)
```

```python
df2["iso_custom"] = (df2["Iso_Score"] < threshold).astype(int)
```

```python
print(classification_report(df2["Class"], df2["iso_custom"]))
```

```python
# recall increased.
```

```python
df_sorted = df2.sort_values("Iso_Score")
```

```python
k = int(len(df_sorted) * 0.01)

top_k = df_sorted.head(k)
```

```python
frauds_captured = top_k["Class"].sum()
total_frauds = df2["Class"].sum()

recall_at_k = frauds_captured / total_frauds

print(recall_at_k)
```

```python
percentages=np.linspace(0.001,0.05,30)
recall=[]
for per in percentages :
    k = int(len(df_sorted) * per)
    captured = df_sorted.head(k)["Class"].sum()
    recall.append(captured / total_frauds)
```

```python
plt.plot(percentages * 100 , recall)
plt.xlabel("investigated transaction")
plt.ylabel("fraud recall")
plt.title("Top-K Capture Curve (Isolation Forest)")

plt.show()
```

```python
from xgboost import XGBClassifier
print("Training with XGBoostClassifier...")
xgb= XGBClassifier()
```

```python
X = df2.drop(["Class", "Iso_Score", "iso_custom"], axis=1)
# anomaly_score is already in X

y = df2["Class"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

xgb = XGBClassifier(
    scale_pos_weight=len(y_train[y_train==0]) / len(y_train[y_train==1]),
    random_state=42, n_jobs=-1
)
#scale_pos_weight to handle unbalanced dataset

xgb.fit(X_train, y_train)
print(classification_report(y_test, xgb.predict(X_test)))
```
